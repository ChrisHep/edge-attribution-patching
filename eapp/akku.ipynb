```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats
import sklearn
from sklearn.linear_model import LinearRegression
```

```python
#old_phone = np.array([94, 91, 85, 80, 75, 69, 67, 66, 63, 57, 56, 52, 49, 46])
#new_phone = np.array([87, 82, 74, 69, 66, 64, 62, 61, 58, 51, 49, 44, 42, 38])
old_phone = np.array([99, 98, 96, 94, 91, 82, 77])
new_phone = np.array([97, 95, 94, 91, 90, 86, 82])
```

```python
plt.plot(old_phone, label='Old Phone')
plt.plot(new_phone, label='New Phone')
plt.legend()
```

```python
X = np.arange(len(old_phone)).reshape(-1, 1)
old_phone_regression = LinearRegression().fit(X, old_phone)
new_phone_regression = LinearRegression().fit(X, new_phone)

print(f"old phone slope: {old_phone_regression.coef_}")
print(f"new phone slope: {new_phone_regression.coef_}")
```
